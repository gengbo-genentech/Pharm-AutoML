```python
import os
cwd = os.getcwd()
print (cwd)
import pandas as pd
import numpy as np
# set up pharmAutoML by "python setup.py install"
from pharmAutoML import autoML_pipeline
import json
```

```python
result_path="./automl_result"
data_dir="./data/risk_factors_cervical_cancer.csv"
target = 'Biopsy'
def load_data(csv_path, target):
    allData=pd.read_csv(csv_path)
    y = allData[target]
    x = allData.drop([target], axis=1)
    return x, y

x, y = load_data(data_dir, target)
para = open('./PharmAutoML/default_parameters.json')
parameter = json.load(para)
# if user want to change parameters, they can change it from here
# parameter['classifier_parameters']['clfs']={'XGboost':{}}
print (parameter)
```

```python
aml=autoML_pipeline.PharmAutoML(x,y)
result_imputeMissing = aml.AutoML(result_path, impute_missing_mode = True, tuning_mode = True, loss_function = 'roc_auc', parameter = parameter)
```
